```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json
```

```python
supers = pd.read_csv('Data/superhero_info - superhero_info.csv')
supers.head()
```

```python
powers = pd.read_csv('Data/superhero_powers - superhero_powers.csv')
powers.head()
```

```python
#split hero/publisher
supers['Hero|Publisher'].str.split('|',expand=True)
```

```python
#add split columns to the hero
supers[['Hero','Publisher']]=supers['Hero|Publisher'].str.split('|',expand=True)
supers.head(2)
```

```python
#drop original column
supers = supers.drop(columns=['Hero|Publisher'])
supers.head(2)
```

```python
#seperate measurments
supers.loc[0,'Measurements']
```

```python
#fix to use json
supers['Measurements']= supers['Measurements'].str.replace("'",'"')
```

```python
supers['Measurements']
```

```python
#json.loads()
supers['Measurements']= supers['Measurements'].apply(json.loads)
supers.loc[0,'Measurements']
```

```python
#make measurements into seperate columns
height_weight = supers['Measurements'].apply(pd.Series)
height_weight
```

```python
#add columns to original
supers = pd.concat((supers,height_weight),axis=1)
supers.head(2)
```

```python
#drop measurements column
supers = supers.drop(columns='Measurements')
supers.head(2)
```

```python
# remove cm from height
supers['Height']= supers['Height'].str.replace('cm','')
supers['Height'].head(2)
```

```python
#remove kg from weight
supers['Weight']= supers['Weight'].str.replace('kg','')
supers['Weight'].head(2)
```

```python
#convert height and weight to numerica
supers['Height']= pd.to_numeric(supers['Height'])
supers['Weight']= pd.to_numeric(supers['Weight'])
```

```python
supers.info()
```

```python
#One hot encode powers
powers['Powers'].head(2)
```

```python
powers.loc[0,'Powers']
```

```python
#split string
powers['powers_split']=powers['Powers'].str.split(',',expand=False)
powers['powers_split']
```

```python
#df explode to create columns
go_boom = powers.explode('powers_split')
```

```python
go_boom
```

```python
#save unique values from exploded
cols_to_make = go_boom['powers_split'].dropna().unique()
cols_to_make
```

```python
#Loop for columns
for col in cols_to_make:
    powers[col]=powers['Powers'].str.contains(col)
powers.head()
```

```python
#merging the dataframes
joined = pd.merge(supers, powers, left_on='Hero',right_on='hero_names')
joined.head()
```

##### Question 1. Compare the average weight of super powers who have Super Speed to those who don't

```python
joined.groupby('Super Speed')['Weight'].mean()
```

###### Superheroes with superspeed are usually heavier/weigh more.

##### Question 2. What is the average height of heros for each publisher?

```python
joined.groupby('Publisher')['Height'].mean()
```
